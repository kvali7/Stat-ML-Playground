### LinearRegression.fit

Fit in OLS solves the following, on training set
Similar to solving a linear Matrix. Has a solution if the gram matrix is non singular
$(X^\top X)$ = gram Matrix 

The dimension of matrix $(X^\top X)$ is p by p and the dimension of vector $y$ is  n

$$
\hat \beta = (X^\top X)^{-1} X^\top y
$$
where $X,y$ are $n \times p$ and $n$ arrays.

1. Matrix multiply
2. Linear solve:
$$
X \hat \beta = y
$$
$$
(X^\top X) \hat \beta = X^\top y
$$

### O notation

- Sequence $A_n = O(B_n)$ if there is an M > 0 such that $A_n \leq M B_n$ for $n$ large enough( $n \geq  n_0$). $O$ is the upper bound.
- Sequence $A_n = \Omega(B_n)$ if there is an M > 0 such that $A_n \geq M B_n$ for $n$ large enough( $n \geq  n_0$). $\Omega$ is the lower bound.
- Sequence $A_n = \Theta(B_n)$ if there is an M > 0 such that $A_n = M B_n$ for $n$ large enough( $n \geq  n_0$).
- An algorithm takes $O(A_n)$ time if the number of basic operations (floating point arithmetic) to run is $O(A_n)$. 

The for loop to compute
$$\sum_{i=1}^n i$$
takes $O(n)$ time while computing
$$n (n + 1)/2$$
takes $O(1)$ time.

### LinearRegression.fit

1. Matrix mult complexity
$$
(X^\top X)_{j,k} = \sum_{i=1}^n X_{i,j} X_{i,k} 
$$
sum is $O(n)$, full complexity is $O(p^2 n)$ time.

1. Matrix solve - Cholesky decomposition - computational complexity $O(p^3)$ time
1. Total complexity of fit: $O(n p^2 + p^3)$.


### LinearRegression.predict

Given $x_{n+1}$,
$$
\hat f(x_{n+1}) = x_{n+1}^\top \hat \beta = \hat \beta_0 + \sum_{j=1}^n \hat \beta_j x_{n+1,j}.
$$
is $O(p)$ time.

#### Conclusions
1. Predict is fast: $O(p)$
2. Fit is slow: $O(np^2)$ (if $n > p$)
3. $X^\top X$ needs to be invertible (impossible if $p > n$ meaning we have high dimesnion!)
4. $\hat f$ is linear (restrictive)

### K nearest neighbors

Given a metric (d), the K-nearest neighbors of $x$ in $x_1,\ldots,x_n$ is $x_{j_1},\ldots,x_{j_K}$ such that
$$
d(x,x_{j_1}) \le d(x,x_{j_2}) \le \ldots
$$

note that $x$ is the test point and $x_1, \ldots, x_n$ are the training points
- KNN methods fit a local model for predicting $x$ using the KNNs
- Intuition is that $Y|X=x$ is approximately iid with $Y|X=x_{j_k}$. We want to learn the distribution of $Y|X=x$.
- This gives us Non-Linear approximation of the data.
- For regression,
$$
\hat f_K(x) = \frac 1K \sum_{k=1}^K y_{j_k}.
$$

### K nearest neighbors.fit

- Naive implementation, save the data
- Better is sort or arrange the data
 - in 1D sort makes nearest neighbor lookup fast!
 - sort takes $O(n \log n)$ time
 - in higher dimensions there is specialized data structions (KDTree)

### K nearest neighbors.predict

- For each new point find the KNNs
 - if unsorted then this can take $O(n)$ time for each new point because we have to compute all the distances and resort - compare to $O(p)$ for OLS
 - if sorted then this can take $O(\log n)$ time (different for $>1$D)
- Take average of $K$ neighbors is fast

We better not touch the test data during training procedure and vice versa

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

class WinePredictKNN:
    def __init__(self, K = 5, selCols = None):
        self.numNeighbors = K
        self.knn = KNeighborsRegressor(n_neighbors = self.numNeighbors)
        self.selCols = selCols
    def standardize(self, X, Scaler = None):
        if Scaler is None:
            self.Scal = StandardScaler()
            self.Scal.fit(X)
        else:
                self.Scal = Scaler
        return self.Scal.transform(X) , self.Scal
    def fit(self, X, y, standardize = False):
        if standardize is True:
            [X, _]  = self.standardize(X)
        if self.selCols is not None:
            X = X[:,self.selCols]
        self.knn.fit(X,y)
    def predict(self, X, standardize = False, Scaler = None):
        if standardize is True:
            if Scaler is None:
                if self.Scal is not None:
                    X = self.Scal.transform(X)
                else:
                    raise ValueError('no Scaler is defined')
            else:
                X = Scaler.transform(X)
        if self.selCols is not None:
            X = X[:,self.selCols]
        return self.knn.predict(X)
```

### Bias-Variance tradeoff for square error loss
#### (STA 251 notes scribed by Jianing Fan, Yuhan Ning)

- square error loss
- any predictor $f$

Risk is
$$
R(f) = \mathbb E \big[(Y - f(X))^2\big].
$$

\begin{align*}
    & \eta(x) = \mathbb{E}[Y|X=x] \quad \text{s.t. minimizes}\, R(f),\\
    & R(\eta) = \mathbb{E}\big[(Y-\eta(X))^2\big]=: R^{*}.
\end{align*}

We cannot get any better that $R^{*}$ because $\eta(X)$ is the best possible estimator.
The term $Y-\eta(X)$ is the irreducable error and $f(X)$ is what we are actually estimating.

\begin{align*}
    \mathbb{E}[(Y-\eta(X))(\eta(X)-f(X))] &= \mathbb{E}\big[\mathbb{E}[(Y-\eta(X))(\eta(X)-f(X))|X]\big]\\
    &= \mathbb{E}\big[\mathbb{E}[(Y-\eta(X))|X](\eta(X)-f(X))\big]\\
    &= \mathbb{E}\left[\mathbb{E}\big[(Y-\mathbb{E}[Y|X])|X\big](\eta(X)-f(X))\right]\\
    &=0.
\end{align*}
This is because $\mathbb{E}\big[(Y-\mathbb{E}[Y|X])|X\big]=\mathbb{E}[Y|X]-\mathbb{E}[Y|X]=0$

$$\mathbb{E}\left[(\hat{f}(X)-\eta(X))^2\right] = \mathbb{E}\left[ \mathbb{E}\big[ \hat{f}(X)-\eta(X))^2 | X \big] \right]$$ 
So let's fix the predicted point $X = x$.

Notice that $f$ is random ($\hat{f}$) because it comes from training data. It's because of the label noise. Let's condition on the training X's ($X_{[n]}$)...

So we have
$$
    R(\hat{f}) = \mathbb{E} \big[(Y-\hat{f}(X))^{2}\big] = R^{*} + \mathbb{E} \big[(\eta(X)-\hat{f}(X))^{2}\big]
$$

$$
R(\hat{f}) = R^* + \mathbb{E}[\text{Bias}^2(\hat{f}(X)|X_{[n]})+\text{Var}(\hat{f}(X)|X_{[n]})].
$$
Risk is irreducible error + bias + variance

where 

\begin{align*}
    \mathbb{E}\left[(\hat{f}(x)-\eta(x))^2\right] &= \mathbb{E}\left[(\hat{f}(x)-\mathbb{E}[\hat{f}(x)|X_{[n]}]+\mathbb{E}[\hat{f}(x)|X_{[n]}]-\eta(x))^2\right]\\
     &= \mathbb{E}\left[(\hat{f}(x)-\mathbb{E}[\hat{f}(x)|X_{[n]}])^2\right]+\mathbb{E}\left[(\mathbb{E}[\hat{f}(x)|X_{[n]}]-\eta(x))^2\right] + (\text{cross-term}=0)\\
    
    &= \mathbb{E}[\text{Var}(\hat{f}(x)|X_{[n]})]+\mathbb{E}[\text{Bias}^2(\hat{f}(x)|X_{[n]})],
\end{align*}

where $\text{Bias}(\hat{f}(x)|X_{[n]}) = \mathbb{E}[\hat{f}(x)|X_{[n]}] -\eta(x)$.

### Exercise 2.2

Recall KNN regression:
$$
\hat f_K(x) = \frac 1K \sum_{k=1}^K y_{j_k(x)}.
$$
where $j_k(x)$ is the index of the kth nearest neighbor of $x$ in $X_1,\ldots,X_n$. (Consider $x$ fixed.)

Assume homoscedasticity: $\text{Var}(y_i) = \sigma^2$ for all i.
1. Derive a simple expression for $\text{Var}(\hat{f}(x)| X_{[n]})$
2. Derive a simple expression for $\text{Bias}(\hat{f}(x)|X_{[n]})$
3. What do you expect the tradeoff to be between these as $K$ gets large?

$$\text{Var}(\hat{f}(x)| X_{[n]}) = \frac{1}{K^2} \sum_{k=1}^K \text{Var}(y_{j_k(x)}| X_{[n]}) = \frac{\sigma^2}{K}$$

$$\text{Bias}(\hat{f}(x)|X_{[n]}) = \mathbb{E}[\hat{f}(x)|X_{[n]}]-\eta(x) = \frac 1K \sum_{k=1}^K \eta(X_{j_k(x)}) - \eta(x)$$

notice that $\mathbb{E}[y_{j_k(x)}| X_{[n]}] =  \eta(X_{j_k(x)})$

1. We can expect that the conditional variance decreases,
2. The bias may increase because $\eta(X_{j_k(x)})$ may be quite different from $\eta(x)$ as $X_{j_k(x)}$ is far from $x$.

### BV tradeoff for KNN

- low K
 - high variance
 - low bias
- high K
 - low variance
 - high bias
