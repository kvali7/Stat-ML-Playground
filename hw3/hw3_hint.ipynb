## Answer to Excerise 1.1

## Answer to Excerise 1.2

The condition for convexity of function $\phi(x) = (\max\{1-x ,0\})^2$ is to have the following $\forall x_1,x_2$:
$$
\phi(ax_1 + (1 - a)x_2) \leq a\phi(x_1) + (1-a)\phi(x_2), \;\; \text{for}\; a \in [0,1]
$$

Now the condition can be in a different form based on disparate ranges for x_1, x_1:

1- $\phi(x_1) = (1-x_1)^2,\, \phi(x_2) = (1-x_2)^2\, \;\;\text{if}\, x_1< 1 , x_2< 1$

2- $\phi(x_1) = 0,\, \phi(x_2) = 0, \;\;\text{if}\, x_1\geq 1 , x_2\geq 1$

3- $\phi(x_1) = 0,\, \phi(x_2) = (1-x_2)^2\, \;\;\text{if}\, x_1\geq 1 , x_2< 1$

4- $\phi(x_1) = (1-x_1)^2,\, \phi(x_2) = 0\, \;\;\text{if}\, x_1< 1 , x_2\geq 1$  

For the first condtition, we will have $\phi^"(x)=2 \geq 0$, So it is convex $\forall x_1,x_2$

For the second condition, we will have $\phi^"(x) = 0 \geq 0$, So it is convex $\forall x_1,x_2$

We can look at the third and the forth condition together. Here if we plug the expressions into the convexity condition, we have:
$$
\begin {align*}
\phi(ax_1+(1-a)x_2) &= \Big( \max\{a(1-x_1)+(1-a)(1-x_2), \, 0\} \Big)^2\\
                    &\leq \Big(\max\{a(1-x_1),\, 0\} \Big)^2\\
                    &= a^2(1-x_1)^2 = a^2\phi(x_1)\\
                    &\leq a\phi(x_1) + (1-a)\phi(x_2)
\end{align*}
$$

Looking at the afformentioned conditions, $\phi(x) = (\max\{1-x ,0\})^2$ is convex $\forall x$.

## Answer to Excerise 1.4
We know that for $\phi(x) = 1-\tanh(\kappa x)\; \forall \kappa >0$ , the first derivative is:
$$
\frac{d}{dx} \phi(x) = - \frac{d}{dx} \tanh(\kappa x)
$$

knowing that $\tanh (x) = \frac{e^x-e^{-x}}{e^x+e^{-x}}$

The first derivative will be:
$$
\begin{align*}
\frac{d}{dx} \phi(x) &= - \frac{d}{dx} \tanh(\kappa x)\\
                     &= - \frac{d}{dx} \Big( \frac{e^{\kappa x}-e^{-\kappa x}}{e^{\kappa x}+e^{-\kappa x}} \Big)\\
                     &= \frac{-4\kappa}{(e^{\kappa x}+e^{-\kappa x})^2}
\end{align*}
$$

Thus the second derivative will be :
$$
\begin{align*}
\frac{d^2}{d^2x} \phi(x) &= - \frac{d^2}{d^2x} \tanh(\kappa x)\\
                     &=  \frac{d}{dx} \frac{-4\kappa}{(e^{\kappa x}+e^{-\kappa x})^2}\\
                     &= \frac{-8\kappa^2 \big( e^{\kappa x}-e^{-\kappa x} \big)}{(e^{\kappa x}+e^{-\kappa x})^3}
\end{align*}
$$

This is not guaranteed to be positive and thus, $\phi(x)$ is not always convex. To be more specific, $\phi(x)$ is not convex $\forall x < 0$.

## Answer to Excerise 2.1

To calculate gradient of the empirical risk we note:
$$
R_n(\beta) = \frac 1n \sum_{i=1}^{n} \ell(y_i,x_i, \beta)
$$
and the gradiant:
$$
\begin{align*}
\nabla_\beta R_n(\beta) &= \nabla_\beta \Big( \frac 1n \sum_{i=1}^{n} \ell(y_i,x_i, \beta) \Big) \\
                        &= \frac 1n \sum_{i=1}^n \nabla_\beta \ell(y_i,x_i, \beta)
\end{align*}
$$

Thus for calculating the gradient of the empirical risk, we should calculate the summed average gradients for each data point $i$. The gradient of $\ell(y_i,x_i,\beta)$ can be calculated in three regions:

1- $y_ix_i^\top \beta < 1$

2- $y_ix_i^\top \beta > 1$

3- $y_ix_i^\top \beta = 1$

For the first region where $\forall x$ we have $y_ix_i^\top \beta < 1$, the loss is $\ell_i(y_i,x_i,\beta) = \phi(y_i x_i^\top \beta) = (\max\{1-y_i x_i^\top \beta,0\})^2 = (1-y_i x_i^\top \beta)^2$

So the gradient of the loss $\nabla_\beta \ell(y_i,x_i,\beta)$ will be:
$$
\nabla_\beta (\phi(y_ix_i^\top \beta)) = -2y_i(1-y_ix_i^\top\beta)x_i)
$$

For the second region where $\forall x$ we have $y_ix_i^\top \beta < 1$, we can easily confirm that the gradient of the loss will be $\nabla_\beta \ell(y_i,x_i,\beta) = 0 $.

for the third point where $y_ix_i^\top \beta = 1$, because the truncated quadratic loss is right continous and monotone nonincreasing, the gradient at $y_ix_i^\top \beta = 1$ will be the same as the gradient at the right side of this point, so :
$$
\nabla_\beta \ell(y_i,x_i,\beta) = \nabla_\beta (\phi(y_ix_i^\top \beta=1)) = \nabla_\beta (\phi(y_ix_i^\top \beta = 1^+)) = 0
$$

Thus the gradient $\forall x$ exists and in summary it is:
$$
\nabla_\beta \ell(y_i,x_i,\beta) 
= \left\{
\begin{array}{lcl}
&-2y_i(1-y_ix_i^\top\beta)x_i &,\textbf{if }y_i x_i^\top \beta< 1 \\
&0 &,\textbf{if } y_i x_i^\top \beta \geq 1
\end{array}
\right.
$$

## Answer to Excerise 2.4

Here when $G(\beta) = \frac 1n \sum_i 2 (x_i x_i^\top 1\{ y_i x_i^\top \beta < 1 \}) + 2 \lambda I$, by assuming the vector $u \in \mathbb{R}^{p}$ and noting that $\lambda > 0$:
$$
\begin{align*}
u^\top G(\beta) u &= \frac 1n \sum_i 2(u^\top x_i x_i^\top u 1\{ y_i x_i^\top \beta < 1 \})+2 \lambda u^\top u \\
                  &= \frac 1n \sum_i 2( \|x_i^\top u\|_2^2 1\{ y_i x_i^\top \beta < 1 \}) + 2 \lambda \| u \|_2^2\\
                  &\geq 0
\end{align*}
$$

Thus the quasi-hesian matrix $G(\beta)$ is Positive semi Definite and inverse for this matrix always exists.

To calculate the optimum estimate of $\hat\beta$ we can use the Quasi-Hesian Matrix and the Gradient Descent algorithm. The psuedo-code will be as such:

1- Initialize the $p$ dimensional vector of initial paramteres as $\beta_{k=0} = 0_{p\times 1}$. Also we might initialize the initail parameters with setting $\beta_{k=0} = \epsilon$ where $\epsilon \in \mathbb{R}^p$ is a vector of inifinitesimal numbers with zero mean and small variance $q$ ($\epsilon \text{~} N(0,q)$).

2- Calculate the emprical risk as well as the quasi-hesian matrix as such for this $\beta_k$:

$$
\nabla_\beta R_n(\beta_k) = \frac 1n \sum_i \big( -2y_i(1-y_ix_i^\top\beta_k)x_i 1\{y_i x_i^\top \beta_k < 1\}+ 2\beta_k \big)
$$
$$
G(\beta_k) = \frac 1n \sum_i 2 (x_i x_i^\top 1\{ y_i x_i^\top \beta_k < 1 \}) + 2 \lambda I
$$
We need to find the inverse of quasi-hesian matrix as well $G(\beta_k)^{-1}$

3- Plug these into the gradient descent expression to find the new parameters estimate $\hat \beta$:

$$
\beta_{k+1} = \beta_k - \big( G(\beta_k) \big)^{-1} \nabla_\beta R_n(\beta_k)
$$

4- Now we need to check if the new estimate improved the estimate at all, and if the improvement of the new paramteres estimate $\beta_{k+1}$ over the current parameters estimate $\beta_k$ is minimal and less than a specifc threshold we can stop the algorithm. To be more specific:

if $ | \beta_{k+1} - \beta_k | > \delta $ where $\delta > 0$ is a small threshold, 

then assign $\beta_k := \beta_{k+1} $ and continue the algotihm from step 2,


else
return with $\hat \beta = \beta_{k+1} $

```python
import numpy as np
```

```python
class DataSimulator:
    """
    Simulate the data for linear classification
    """
    def __init__(self,p,noise_dist = "normal"):
        self.beta = np.random.normal(0,1,p)
        self.noise_dist = noise_dist
        self.p = p
        
    def sim_data(self,m = 1):
        p = self.p
        X = np.random.normal(0,1,(m,p))
        if self.noise_dist == "normal":
            eps = np.random.normal(0,1,m)
        if self.noise_dist == "chisquare":
            eps = np.random.chisquare(1,m)
        z = X @ self.beta + eps
        y = 2*(z > 0)-1
        return X, y
```

```python
ds = DataSimulator(p=10)
X, y = ds.sim_data(m=10)
print (X.shape, y.shape)
```

```python
def grad_descent(ds,T,m,beta_start,C,gamma):
    """
    ds: data simulator
    T: number of iterations
    m: minibatch size
    beta_start: init beta
    C, gamma : eta_t = C t^(-gamma)
    """
    beta = beta_start
    p = beta_start.shape[0]
    losses = [] # tracks loss at each time
    betas = [] # tracks betas at each time
    deltas = [] # tracks deltas at each time
    beta_true = ds.beta
    # loop over iterations
    for t in range(1, T+1):
        X_, y_ = ds.sim_data(m= m)

        # update beta by gradient descent over truncated quadratic loss
        stepsize = C * t **(-gamma)
        phi = np.maximum(1 - y_ * (X_ @ beta), np.zeros(m))
        gradient_risk = -(2/m) * (np.array(y_ * phi ) @ X_.reshape(m, p))
        beta_ = beta - stepsize * gradient_risk
        # append the new beta to betas list
        betas.append(beta_)

        #calculate squared error loss for the new beta and append to losses list
        loss = sum((beta_ - beta_true)**2)
        losses.append(loss)

        # find how much the parameters have improved 
        delta = np.sqrt(sum((beta_ - beta)**2))
        deltas.append(delta)

        # replace the current beta with the new beta to resume the loop
        beta = beta_

    return(betas,losses,deltas)
```

To ensure that we are actually comparing these three together, we will modify the function, so that the losses for all three learning scheduling will be calculated at once. Because the batches should be the same in the comparison and to avoid incorrect comparison.

```python
def grad_descents(ds,T,m,beta_start,C,gamma):
    """
    ds: data simulator
    T: number of iterations
    m: minibatch size
    beta_start: init beta
    C, gamma : eta_t = C t^(-gamma)
    """
    beta_1 = beta_start
    beta_2 = beta_start
    beta_3 = beta_start
    p = beta_start.shape[0]
    losses_1 = [] # tracks loss at each time for first learning schedule
    betas_1 = [] # tracks betas at each time for first learning schedule
    losses_2 = [] # tracks loss at each time for second learning schedule
    betas_2 = [] # tracks betas at each time for second learning schedule
    losses_3 = [] # tracks loss at each time for third learning schedule 
    betas_3 = [] # tracks betas at each time for third learning schedule 
    beta_true = ds.beta
    # loop over iterations
    for t in range(1, T+1):
        X_, y_ = ds.sim_data(m= m)

        # for first scheduling
        # update beta by gradient descent over truncated quadratic loss
        stepsize = C * t **(-gamma[0])
        phi = np.maximum(1 - y_ * (X_ @ beta_1), np.zeros(m))
        gradient_risk = -(2/m) * (np.array(y_ * phi ) @ X_.reshape(m, p))
        beta_1_ = beta_1 - stepsize * gradient_risk
        # append the new beta to betas list
        betas_1.append(beta_1_)
        #calculate squared error loss for the new beta and append to losses list
        loss = sum((beta_1_ - beta_true)**2)
        losses_1.append(loss)
        # replace the current beta with the new beta to resume the loop
        beta_1 = beta_1_

        # for second scheduling
        # update beta by gradient descent over truncated quadratic loss
        stepsize = C * t **(-gamma[1])
        phi = np.maximum(1 - y_ * (X_ @ beta_2), np.zeros(m))
        gradient_risk = -(2/m) * (np.array(y_ * phi ) @ X_.reshape(m, p))
        beta_2_ = beta_2 - stepsize * gradient_risk
        # append the new beta to betas list
        betas_2.append(beta_2_)
        #calculate squared error loss for the new beta and append to losses list
        loss = sum((beta_2_ - beta_true)**2)
        losses_2.append(loss)
        # replace the current beta with the new beta to resume the loop
        beta_2 = beta_2_

        # for third scheduling
        # update beta by gradient descent over truncated quadratic loss
        stepsize = C * t **(-gamma[2])
        phi = np.maximum(1 - y_ * (X_ @ beta_3), np.zeros(m))
        gradient_risk = -(2/m) * (np.array(y_ * phi ) @ X_.reshape(m, p))
        beta_3_ = beta_3 - stepsize * gradient_risk
        # append the new beta to betas list
        betas_3.append(beta_3_)
        #calculate squared error loss for the new beta and append to losses list
        loss = sum((beta_3_ - beta_true)**2)
        losses_3.append(loss)
        # replace the current beta with the new beta to resume the loop
        beta_3 = beta_3_

    return(betas_1, losses_1, betas_2, losses_2, betas_3, losses_3)

```

```python
import nltk
nltk.download()
# Download Corpora -> stopwords, Models -> punkt

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
def parse_reu(reu):
    """Parses the etree object and returns a list of dictionary of reuters attr
    Output: {'topics': the topic of the article, 'places': where it is located, 
        'split': training/test split, 'body':the text of the article as a set of words with stopwords removed}
    """
    root= reu.getroot()
    articles = root.body.getchildren()
    stop_words = set(stopwords.words('english'))
    reu_pl = []
    for a in articles:
        reu_parse = {}
        if a.attrib['topics'] != 'YES':
            next
        topics = a.find('topics').findall('d')
        if topics:
            reu_parse['topics'] = [t.text for t in topics]
        else:
            reu_parse['topics'] = []
        places = a.find('places').findall('d')
        if places:
            reu_parse['places'] = [t.text for t in places]
        reu_parse['split'] = a.attrib['lewissplit']
        rtxt = a.find('text')
        word_tokens = word_tokenize(rtxt.text_content())
        filtered_sentence = set([w.lower() for w in word_tokens if not w in stop_words])
        reu_parse['body'] = filtered_sentence
        reu_pl.append(reu_parse)
    return reu_pl
```

```python
reu_pl = parse_reu(reu)
```
